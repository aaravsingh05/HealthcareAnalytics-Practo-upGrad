Analyse Health and Demographic Data to Identify Common Traits Leading to Heart Disease

Part 1 :  Data Preparation and Cleaning

Importing Necessary Libraries


```python
import pandas as pd
```

Importing the Dataset

```python
data = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")
```

Converting the Data into a Dataframe

```python
df = pd.DataFrame(data)
```

Reviewing the Dataset

```python
#First 5 Rows of the Dataset

print(df.head())
```

```python
#Information about the dataset

print(df.info())
```

```python
#Number of Rows and Columns

print(df.shape)
```

```python
#Check for empty values

print(df.isnull().sum())
```

```python
# Number of duplicate rows

print(df.duplicated().sum())
```

```python
#Handling Duplicate Values

df = df.drop_duplicates()
```

Part 2 : Exploratory Data Analysis

Importing Necessary Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Section 1 : Univariate Analysis

1.1 :
Examining Distributions of Individual Variables

```python
#Histogram for distribution of Age

age_groups = {
    1: "18–24",
    2: "25–29",
    3: "30–34",
    4: "35–39",
    5: "40–44",
    6: "45–49",
    7: "50–54",
    8: "55–59",
    9: "60–64",
    10: "65–69",
    11: "70–74",
    12: "75–79",
    13: "80+"
}
print("The Different Age Groups are:\n", age_groups)
sns.histplot(df['Age'], bins= 13, color = 'skyblue')
plt.title("Age Group Distributions")
plt.xlabel("Age Group")
plt.show()
```

```python
#Distribution of Body Mass Index(BMI)

sns.histplot(df['BMI'], bins= 13, color = "salmon")
plt.title("BMI Distribution")
plt.xlabel("Body Mass Index")
plt.show()
```

```python
#Distribution of Mental Heath

sns.histplot(df['MentHlth'], bins=13, color='lightgreen')
plt.title("Mental Health Days (Past 30 Days)")
plt.xlabel("Days Mentally Unwell")
plt.show()
```

```python
#Distribution of Physical Health

sns.histplot(df['PhysHlth'], bins=13, color='orchid')
plt.title("Physical Health Days (Past 30 Days)")
plt.xlabel("Days Physically Unwell")
plt.show()
```

1.2 : Investigating Prevalance of Health Condition

```python
# Renaming Diabetes for better clarity

df['Diabetes'] = df['Diabetes'].replace({0: 'No Diabetes', 1: 'Pre-Diabetes', 2: 'Diabetes'})
```

```python
# List of binary condition variables to plot

condition_vars = {
    'HighBP': 'High Blood Pressure',
    'HighChol': 'High Cholesterol',
    'Smoker': 'Smoker (Ever Smoked 100+ Cigarettes)',
    'Stroke': 'Ever Had a Stroke',
    'PhysActivity': 'Physical Activity (Past 30 Days)',
    'DiffWalk': 'Difficulty Walking',
    'HvyAlcoholConsump': 'Heavy Alcohol Consumption',
}
```

```python
#Different Health Concerns

for var, label in condition_vars.items():
    plt.figure(figsize=(6, 4))
    sns.countplot(x=var, hue=var, data=df, palette=["lightgreen", "red"], legend=False)
    plt.title(f'Prevalence of {label}')
    plt.xticks([0, 1], ['No', 'Yes'])
    plt.xlabel(label)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

#Presence of Diabetes (3 Levels)

plt.figure(figsize=(6, 4))
sns.countplot(x='Diabetes', hue='Diabetes', data=df, palette=["lightgreen", "yellow", "red"], legend=False)
plt.title('Diabetes Status')
plt.xlabel('Diabetes Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

1.3 : Analysing the Distribution of Heart Disease

```python
# Bar Chart

plt.figure(figsize=(6, 4))
sns.countplot(x= 'HeartDiseaseorAttack', hue= 'HeartDiseaseorAttack', data=df, palette=["lightgreen", "red"], legend=False)
plt.title('Distribution of Heart Disease / Heart Attack')
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Heart Disease or Heart Attack')
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart

plt.figure(figsize=(5, 5))
labels = ['No Heart Disease', 'Heart Disease']
sizes = df['HeartDiseaseorAttack'].value_counts().sort_index()
colors = ['lightgreen', 'red']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Heart Disease / Heart Attack Proportion')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Section 2 : Bivariate Analysis

2.1 : Exploring Relationships with Heart Disease

```python
# Defining key variables to compare

key_vars = {
    'HighBP': 'High Blood Pressure',
    'HighChol': 'High Cholesterol',
    'Smoker': 'Smoker',
    'Stroke': 'Had a Stroke',
    'PhysActivity': 'Physical Activity (Past 30 Days)',
    'DiffWalk': 'Difficulty Walking',
    'HvyAlcoholConsump': 'Heavy Alcohol Consumption',
}
```

```python
# Bar Chart for Binary Health Condition vs Heart Disease

for var, label in key_vars.items():
    sns.countplot(x=var, hue='HeartDiseaseorAttack', data=df, palette=['lightgreen', 'red'])
    plt.title(f"{label} vs Heart Disease")
    plt.xlabel(label)
    plt.ylabel('Count')
    plt.xticks([0, 1], ['No', 'Yes'])
    plt.legend(title="Heart Disease", labels=['No', 'Yes'])
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot for BMI by Heart Disease

sns.boxplot(x = 'HeartDiseaseorAttack', y = 'BMI', data = df, color = "lightblue")
plt.title('BMI Distribution by Heart Disease')
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.xlabel('Heart Disease Status')
plt.ylabel('BMI')
plt.tight_layout()
plt.show()
```

2.2 : Visualising Correlations Between Variables

```python
# Selecting Continous Variables 

cont_vars = ['Age', 'BMI', 'MentHlth', 'PhysHlth', 'GenHlth']
```

```python
# Computing Correlation Matrix

corr_matrix = df[cont_vars].corr()
```

```python
# Plotting the Heatmap

sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm", linewidths = 0.5)
plt.title("Correlation Matrix Between Continous Variables")
plt.tight_layout()
plt.show()
```

2.3 : Compare Heart Diseases Across Various Demographic Groups

```python
#Heart Diseases by Age

plt.figure(figsize=(10,5))
sns.countplot(x='Age', hue='HeartDiseaseorAttack', data=df, palette= ['lightgreen', 'red'])
plt.title("Heart Disease by Age Group")
plt.xlabel("Age Group Code (1 = 18–24, ..., 13 = 80+)")
plt.ylabel("Count")
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
#Heart Diseases by Gender

plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='HeartDiseaseorAttack', data=df, palette= ['lightgreen', 'red'])
plt.title("Heart Disease by Gender")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
#Heart Diseases by Education Level

edu_levels = { 1: "No School",
    2: "Elementary",
    3: "Some High School",
    4: "High School Grad",
    5: "Some College",
    6: "College Grad"
    }
print(f"Education Levels: {edu_levels}")

plt.figure(figsize=(10,5))
sns.countplot(x='Education', hue='HeartDiseaseorAttack', data=df, palette= ['lightgreen', 'red'])
plt.title("Heart Disease by Education Level")
plt.xlabel("Education")
plt.ylabel("Count")
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

Part 3 : Data Visualisation

Section 1 : Bar Plots and Pie Charts

1.1 : Visualising Catagorical Variables

```python
#Defining Catagorical Variables 

cat_var = {
    'PhysActivity' : 'Physical Activity(Part 30 Days)',
    'Fruits' : 'Fruits Consumption', 
    'Veggies' : 'Vegetable Consumption'
}
```

```python
#Plotting the Bar Chart

for var, label in cat_var.items():
    sns.countplot(x = var, data = df, palette = ['red', 'lightgreen'], hue = var, legend= False)
    plt.title(label)
    plt.xticks([0, 1], ['No', 'Yes'])
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

1.2 : Compare Heart Diseases Across Catagories

```python
#Catagorical Variables for Grouped Bar Chart

cat_vars = {
    'Smoker': 'Smoking Status',
    'PhysActivity': 'Physical Activity',
    'Fruits': 'Fruit Consumption',
    'Veggies': 'Vegetable Consumption',
    'Sex': 'Gender',
    'Education': 'Education Level'
}
```

```python
#Creating the Grouped Bar Charts
 
for var, label in cat_vars.items():
    plt.figure(figsize=(6, 4))
    sns.countplot(x=var, hue='HeartDiseaseorAttack', data=df, palette=['lightgreen', 'red'])
    plt.title(f"Heart Disease Prevalence by {label}")
    plt.xlabel(label)
    plt.ylabel("Count")
    plt.legend(title='Heart Disease', labels=['No', 'Yes'])
    plt.xticks([0, 1], ['No', 'Yes']) 
    plt.tight_layout()
    plt.show()
```

Section 2 : Other Visualisation Techniques

2.1 : Displaying Distribution of Continous Variable

```python
cont_vars = ['BMI', 'MentHlth', 'PhysHlth']
```

```python
# Histograms

for var in cont_vars:
    plt.figure(figsize=(6,4))
    sns.histplot(df[var], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Boxplots

for var in cont_vars:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[var], color='lightcoral')
    plt.title(f'Boxplot of {var}')
    plt.xlabel(var)
    plt.tight_layout()
    plt.show()
```

2.2 : Heatmap and Correlation Plots

```python
# Select continuous and ordinal variables

corr_vars = ['BMI', 'MentHlth', 'PhysHlth', 'GenHlth', 'Age']
```

```python
#Plotting the Correlation Matrix

corr_matrix = df[corr_vars].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

2.3 : Scatterplots

```python
#Scatterplot for BMI vs Mental Health

#Plot the Scatterplot
sns.scatterplot(data=df, x='BMI', y='MentHlth', alpha=0.5, palette= 'husl', hue = 'HeartDiseaseorAttack')
plt.title("BMI vs Mental Health Days")
plt.xlabel("BMI")
plt.ylabel("MentHlth")
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

2.4 : Line Plots

```python
#Line Plot of Age vs BMI
age_bmi = df.groupby('Age')['BMI'].mean().reset_index()

plt.figure(figsize=(7,4))
sns.lineplot(data=age_bmi, x='Age', y='BMI', marker='o')
plt.title("Average BMI Across Age Groups")
plt.xlabel("Age Group (1 = 18–24, ..., 13 = 80+)")
plt.ylabel("Average BMI")
plt.tight_layout()
plt.show()
```

2.5 : Violin Plot

```python
#Plotting the Violin Plot

plt.figure(figsize=(6,4))
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette= ['lightgreen', 'red'], hue = 'HeartDiseaseorAttack', legend= False)
plt.title("BMI Distribution by Heart Disease Status")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("BMI")
plt.tight_layout()
plt.show()
```

2.6 : Pair Plot

```python
#Subsets for the Pair Plot
subset = df[['BMI', 'MentHlth', 'PhysHlth', 'GenHlth', 'HeartDiseaseorAttack']]

#Plotting the Pair Plot
sns.pairplot(subset, hue='HeartDiseaseorAttack', palette='husl', diag_kind='kde')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.tight_layout()
plt.show()
```
